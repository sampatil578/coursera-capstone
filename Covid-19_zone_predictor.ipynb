```python
df = pd.read_csv('covid-19.csv')
```

```python
map_india = folium.Map(location=[latitude, longitude], zoom_start=4)

# add markers to map
for lat, lng, place_name,zone in zip(df['Latitude'], df['Longitude'], df['Place Name'],df['Zone']):
    label = folium.Popup(place_name, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=2,
        popup=label,
        color=zone,
        fill=True,
        fill_color=zone,
        fill_opacity=0.7,
        parse_html=False).add_to(map_india)  
    
map_india
```

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
from sklearn import preprocessing
%matplotlib inline
```

```python
X= preprocessing.StandardScaler().fit(X).transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.metrics import jaccard_similarity_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
```

# K Nearest Neighbor(KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
Ks = 15
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))
ConfustionMx = [];
for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

mean_acc
```

```python
k = 1
#Train Model and Predict  
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
yhat6 = neigh.predict(X_test)
v1 = f1_score(y_test, yhat6, average='weighted')
v2 = jaccard_similarity_score(y_test, yhat6)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
max_d = 3
ks = 15
mean_ax = np.zeros((Ks-3))
for max_d in range(3,ks):
    drugTree = DecisionTreeClassifier(criterion="entropy", max_depth = max_d)
    drugTree.fit(X_train,y_train)
    predTree = drugTree.predict(X_test)
    mean_ax[max_d-3] = metrics.accuracy_score(y_test, predTree)
mean_ax
```

```python
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth = 12
                                 )
drugTree.fit(X_train,y_train)
predTree = drugTree.predict(X_test)

v3 = f1_score(y_test, predTree, average='weighted')
v4 = jaccard_similarity_score(y_test, predTree)
```

# Support Vector Machine

```python
from sklearn import svm
clf = svm.SVC(kernel='rbf')
clf.fit(X_train, y_train) 
yhatsvm = clf.predict(X_test)
```

```python
v5 = f1_score(y_test, yhatsvm, average='weighted')
v6 = jaccard_similarity_score(y_test, yhatsvm)
```

# Logistic Regreession

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR
```

```python
yhat = LR.predict(X_test)
yhat_prob = LR.predict_proba(X_test)
v7 = f1_score(y_test, yhat, average='weighted')
v8 = jaccard_similarity_score(y_test, yhat)
v9 = log_loss(y_test, yhat_prob)
```

# Model Evaluation

```python
# initialize list of lists 
data = [['KNN',v1,v2,np.nan],['Decision',v3,v4,np.nan],['SVM',v5,v6,np.nan],['LogisticRegression',v7,v8,v9]] 
  
# Create the pandas DataFrame 
df_final = pd.DataFrame(data, columns = ['Algorithm','Jaccard','F1-score','LogLoss']) 
  
# print dataframe. 
df_final 
```
